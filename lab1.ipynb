```python
import numpy as np
import pandas as p
import matplotlib.pyplot as plt



list1= [1,2,3,4]
list2 = [6,7,8,9]
print(list1)
print(list2)

array1=np.array([1,2,3,4])
array2=np.array([5,6,7,8])
print(array1)
print(array2)

print(type(list1))
print(type(list2))
print(type(array1))
print(type(array2))

print("multiplication:", (array1* array2))
print("division:", (array1/array2))
print("poweroff:", (array1** array1+array2))

print(np.sin(array1))
print(np.log(array1))
print(np.log2(array1))
print(np.exp(array1))

people = ["kiran", "arun", "vijay", "varun"]
age =[25, 30, 35, 40,]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]

plt.figure(figsize=(6, 6))
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='red', label='Weight')
plt.title('Scatter Plot of Age vs. Height and Weight')
plt.xlabel('Age')
plt.ylabel('Measure')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
bar_width = 0.3
index = range(len(people))
plt.bar(index, height, width=bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in index], weight, width=bar_width, color='red', label='Weight')
plt.xlabel('People')
plt.ylabel('Measure')
plt.title('Bar Chart of Height and Weight')
plt.xticks([i + bar_width / 2 for i in index], people)
plt.legend()
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(height, bins=5, alpha=0.7, color='blue', label='Height')
plt.hist(weight, bins=5, alpha=0.7, color='red', label='Weight')
plt.xlabel('Measure')
plt.ylabel('Frequency')
plt.title('Histogram of Height and Weight')
plt.legend()
plt.grid(True)
plt.show()
```
